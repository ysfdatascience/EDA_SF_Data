```python
import pandas as pd
store = pd.HDFStore(path = r"C:\Users\yy\Desktop\new_phase\python\dataset\store.h5")
```

```python
df = store["/SF_salary"];df.head()
```

```python
df.columns = list(map(lambda x: x.lower(), df.columns));df.head()
```

```python
df.info()
```

```python
def transform_series(item):
    result = pd.to_numeric(df[item], errors="coerce")
    result = result.fillna(0)
    return result
    

for item in ["overtimepay", "benefits", "otherpay", "basepay"]:
    df[item] = transform_series(item)
```

<h1>find the employees who earn the highest totalpaybenefits(all the payment employee got) amount in each year.</h1>

<h2>Solution 1</h2>

```python
year_list = df.year.unique().tolist();year_list
```

First, extract data for year 2011, sort the table and take the first row

```python
df[df["year"] == 2011].sort_values(by = "totalpaybenefits", ascending = False).iloc[0].to_frame().T
```

Now we write the above script in for loop for each year

```python
new_df = pd.DataFrame(columns = df.columns);
for item in year_list:
    new_df = pd.concat([new_df, df[df["year"] == item].sort_values(by = "totalpaybenefits", ascending = False).iloc[0].to_frame().T], ignore_index=True)

new_df
```

<h1>Find the number of unique job titles and print 10 most frequent job titles. </h1>

```python
f"there are {df.jobtitle.nunique()} items in jobtitle variable"
```

```python
df.jobtitle.value_counts().nlargest(n = 10)
```

Calculate the average TotalPay values for each of the the 10 most frequent job titles

```python
df[df.jobtitle.isin(df.jobtitle.value_counts().nlargest(n = 10).index)].groupby(by = "jobtitle")[["totalpay"]].mean().sort_values(by = "totalpay", ascending = False).rename(columns = {"totalpay": "avg_total_pay"}).style.format(decimal = ".", thousands = ",", precision = 2)
```

```python
time_df = df.copy()
```

```python
time_df.set_index("year", inplace = True)
```

```python
time_df = time_df.select_dtypes(include = ["float", "int"]).groupby("year").agg(["min", "mean", "max", "std"])
```
